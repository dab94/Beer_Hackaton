```python
#import machine learning related libraries
from sklearn.svm import SVC
from sklearn.multioutput import MultiOutputClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
import xgboost as xgb
from sklearn.model_selection import KFold, cross_val_score, train_test_split
```

```python
df_test = pd.read_csv('./reto-colombia-2020-main/reto-colombia-2020-main/data/Input3_clientes_test.csv', sep=';')
df_test["Date"] = "2020-10"
#df_test['Date'] = df_test['Date'].apply(lambda x: x.strftime('%Y-%m'))
df_test.set_index('Date', inplace=True)
X_test = df_test["Cliente"].to_frame()
```

```python
df_marca2 =df.groupby(['Date','Cliente'])['Marca2'].sum().reset_index()
df_marca2.set_index('Date', inplace=True)


X_train,Y_train = df_marca2.drop("Marca2", axis=1),df_marca2['Marca2']
xgb_model = xgb.XGBClassifier(max_depth=11, min_child_weight=5).fit(X_train,Y_train)
probs = xgb_model.predict_proba(X_test)
probs = probs[:, 1]
df_test["Marca2"] = probs.round(4)


df_marca1 = df.groupby(['Date','Cliente'])['Marca1'].sum().reset_index()
df_marca1.set_index('Date', inplace=True)
df_marca1 = df_marca1.loc["2020-07":]
df_marca1
X_train,Y_train = df_marca1.drop("Marca1", axis=1),df_marca1['Marca1']
xgb_model = xgb.XGBClassifier(max_depth=11, min_child_weight=5).fit(X_train,Y_train)
probs = xgb_model.predict_proba(X_test)
probs = probs[:, 1]
df_test["Marca1"] = probs.round(4)

df_marca3 =df.groupby(['Date','Cliente'])['Marca3'].sum().reset_index()
df_marca3.set_index('Date', inplace=True)
X_train,Y_train = df_marca3.drop("Marca3", axis=1),df_marca3['Marca3']
xgb_model = xgb.XGBClassifier(max_depth=11, min_child_weight=5).fit(X_train,Y_train)
probs = xgb_model.predict_proba(X_test)
probs = probs[:, 1]
df_test["Marca3"] = probs.round(4)


df_marca_inno1 = df.groupby(['Date','Cliente'])['Marca_Inno1'].sum().reset_index()
df_marca_inno1.set_index('Date', inplace=True)
df_marca_inno1 = df_marca_inno1.loc["2020-07":]
df_marca_inno1
X_train,Y_train = df_marca_inno1.drop("Marca_Inno1", axis=1),df_marca_inno1['Marca_Inno1']
xgb_model = xgb.XGBClassifier(max_depth=11, min_child_weight=5).fit(X_train,Y_train)
probs = xgb_model.predict_proba(X_test)
probs = probs[:, 1]
df_test["Marca_Inno1"] = probs.round(4)

df_marca_inno2 = df.groupby(['Date','Cliente'])['Marca_Inno2'].sum().reset_index()
df_marca_inno2.set_index('Date', inplace=True)
df_marca_inno2 = df_marca_inno2.loc["2020-07":]
df_marca_inno2
X_train,Y_train = df_marca_inno2.drop("Marca_Inno2", axis=1),df_marca_inno2['Marca_Inno2']
xgb_model = xgb.XGBClassifier(max_depth=11, min_child_weight=5).fit(X_train,Y_train)
probs = xgb_model.predict_proba(X_test)
probs = probs[:, 1]
df_test["Marca_Inno2"] = probs.round(4)
```
